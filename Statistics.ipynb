# Descriptive Statistics

```python
# Example dataset - students' exam scores
exam_scores = np.array([85, 78, 92, 70, 88, 82, 90, 95, 76, 84])

# Compute the mean
mean_score = np.mean(exam_scores)

# Compute the median
median_score = np.median(exam_scores)

# Compute the mode
mode_score = np.argmax(np.bincount(exam_scores))

# Compute the variance
variance_score = np.var(exam_scores)

# Compute the standard deviation
std_deviation_score = np.std(exam_scores)

# Compute the range
range_score = np.max(exam_scores) - np.min(exam_scores)

# Compute the interquartile range (IQR)
q1 = np.percentile(exam_scores, 25)
q3 = np.percentile(exam_scores, 75)
iqr = q3 - q1

# Compute the skewness
skewness_score = stats.skew(exam_scores)

# Compute the kurtosis
kurtosis_score = stats.kurtosis(exam_scores)

# Compute the percentiles
percentiles_scores = np.percentile(exam_scores, [25, 50, 75])

# Compute the count
count_scores = len(exam_scores)

# Compute the sum
sum_scores = np.sum(exam_scores)

# Print the results
print("Exam Scores: ", exam_scores)
print("Mean: ", mean_score)
print("Median: ", median_score)
print("Mode: ", mode_score)
print("Variance: ", variance_score)
print("Standard Deviation: ", std_deviation_score)
print("Range: ", range_score)
print("Interquartile Range (IQR): ", iqr)
print("Skewness: ", skewness_score)
print("Kurtosis: ", kurtosis_score)
print("25th, 50th, and 75th Percentiles: ", percentiles_scores)
print("Count: ", count_scores)
print("Sum: ", sum_scores)
```

# Outliers Detection

```python
# Create a sample dataset of students' exam scores
data = {
    'StudentID': range(1, 11),
    'ExamScore': [85, 90, 78, 92, 88, 200, 85, 80, 95, 87]
}

# Create a DataFrame from the data
df = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset:")
print(df)

# Find and remove outliers using the IQR method
Q1 = df['ExamScore'].quantile(0.25)
Q3 = df['ExamScore'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers_mask = (df['ExamScore'] < lower_bound) | (df['ExamScore'] > upper_bound)

# Identify and display outliers
outliers = df[outliers_mask]
print("\nOutliers:")
print(outliers)

# Create a box plot to visualize the distribution and outliers
plt.figure(figsize=(8, 6))
plt.boxplot(df['ExamScore'], vert=False)
plt.xlabel('Exam Scores')
plt.title('Box Plot of Exam Scores')
plt.show()
```

In the boxplot above, we can see that there is one outlier, i.e. 200.

# Outliers Removal

```python
# Remove outliers from the dataset
cleaned_df = df[~outliers_mask]

# Display the cleaned dataset
print("\nCleaned Dataset (Outliers Removed):")
print(cleaned_df)

# Create a box plot to visualize the distribution and outliers removal
plt.figure(figsize=(8, 6))
plt.boxplot(cleaned_df['ExamScore'], vert=False)
plt.xlabel('Exam Scores')
plt.title('Box Plot of Exam Scores')
plt.show()
```

# Confidence Interval

You have a dataset of customer wait times in a retail store. The store manager wants to estimate the average wait time for all customers in the store and determine a 95% confidence interval around that estimate. This will help the manager make decisions about staffing and customer service improvements.

```python
# Wait time data (in minutes)
wait_times = [3, 5, 6, 4, 7, 2, 5, 6, 8, 4, 5, 3, 6, 4, 7, 5, 6, 3, 4, 5]

# Step 1: Calculate the sample mean and standard error
sample_mean = np.mean(wait_times)
standard_error = np.std(wait_times, ddof=1) / np.sqrt(len(wait_times))

# Step 2: Set the confidence level and find the critical value
confidence_level = 0.95
alpha = 1 - confidence_level
critical_value = stats.t.ppf(1 - alpha / 2, df=len(wait_times) - 1)

# Step 3: Calculate the confidence interval
lower_bound = sample_mean - critical_value * standard_error
upper_bound = sample_mean + critical_value * standard_error

# Step 4: Interpret the confidence interval
print(f"The {confidence_level*100}% confidence interval for the average wait time is between {lower_bound:.2f} and {upper_bound:.2f} minutes.")
```

Based on the sample data, we are 95% confident that the true average wait time for all customers in the retail store lies within the interval 4.17 and 5.63 minutes. This means that if we were to repeatedly take samples of wait times and calculate confidence intervals, approximately 95% of those intervals would contain the true average wait time.

With this information, the store manager can make informed decisions about staffing levels during peak hours and identify opportunities to improve customer service by reducing wait times.

#  Hypothesis testing using z-score

**Scenario:** A company that produces energy drinks wants to test whether a new formulation of their drink results in a significant increase in the average sales volume compared to the old formulation.

**Null Hypothesis (H0):** The new formulation has no significant effect on the average sales volume.

**Alternative Hypothesis (H1):** The new formulation results in a significant increase in the average sales volume.

```python
# Sample data (sales volumes in units)
old_formulation = np.array([120, 140, 100, 130, 160, 110, 135, 125, 115, 130,
                            120, 125, 130, 115, 110, 140, 120, 125, 130, 115,
                            150, 140, 110, 135, 125, 130, 115, 120, 140, 100,
                            130, 160, 110, 135, 125, 115, 130, 120, 125, 130,
                            115, 150, 140, 110, 135, 125, 130, 115, 120, 140])

new_formulation = np.array([160, 150, 140, 155, 145, 150, 140, 165, 150, 155,
                            160, 145, 150, 140, 155, 145, 150, 140, 165, 150,
                            160, 150, 140, 155, 145, 150, 140, 165, 150, 155,
                            160, 145, 150, 140, 155, 145, 150, 140, 165, 150,
                            160, 150, 140, 155, 145, 150, 140, 165, 150, 155])
```

```python
# Calculate the sample means and standard deviations
mean_old = np.mean(old_formulation)
mean_new = np.mean(new_formulation)

std_dev_old = np.std(old_formulation, ddof=1)  # Use ddof=1 for sample standard deviation
std_dev_new = np.std(new_formulation, ddof=1)

# Calculate the standard error of the mean difference
se_diff = np.sqrt((std_dev_old**2 / len(old_formulation)) + (std_dev_new**2 / len(new_formulation)))

# Calculate the Z-score
Z = (mean_new - mean_old) / se_diff

# Define the significance level (alpha) - let's use 0.05
alpha = 0.05

# Calculate the critical Z-value for a two-tailed test at alpha level
critical_z = stats.norm.ppf(1 - alpha / 2)

# Check if Z falls within the critical region
if abs(Z) > critical_z:
    print("Reject the null hypothesis. There is a significant increase in average sales volume.")
else:
    print("Fail to reject the null hypothesis. There is no significant increase in average sales volume.")
```

# Hypothesis testing using t-score

**Scenario:** A company that manufactures smartphones wants to test whether a new manufacturing process results in a significant difference in the battery life of their smartphones compared to the old process.

**Null Hypothesis (H0):** The new manufacturing process has no significant effect on the battery life of the smartphones.

**Alternative Hypothesis (H1):** The new manufacturing process results in a significant difference in the battery life of the smartphones.

```python
# Sample data (battery life in hours)
old_process = np.array([24, 23, 26, 22, 21, 23, 25, 24, 23, 22,
                        21, 24, 22, 23, 26, 24, 25, 22, 23, 21])

new_process = np.array([28, 27, 25, 29, 26, 28, 27, 25, 26, 29,
                        27, 28, 28, 27, 25, 26, 28, 29, 27, 28])
```

```python
# Calculate the sample means and standard deviations
mean_old = np.mean(old_process)
mean_new = np.mean(new_process)

std_dev_old = np.std(old_process, ddof=1)  # Use ddof=1 for sample standard deviation
std_dev_new = np.std(new_process, ddof=1)

# Calculate the standard error of the mean difference
se_diff = np.sqrt((std_dev_old**2 / len(old_process)) + (std_dev_new**2 / len(new_process)))

# Calculate the t-score
t_score = (mean_new - mean_old) / se_diff

# Define the degrees of freedom
df = len(old_process) + len(new_process) - 2

# Define the significance level (alpha) - let's use 0.05
alpha = 0.05

# Calculate the critical t-value for a two-tailed test at alpha level
critical_t = stats.t.ppf(1 - alpha / 2, df)

# Check if the t-score falls within the critical region
if abs(t_score) > critical_t:
    print("Reject the null hypothesis. There is a significant difference in battery life.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in battery life.")
```

# Hypothesis testing using ANOVA

**Scenario:** A company that produces three different types of chocolates wants to test whether there is a significant difference in the average sweetness level among these three types.

**Null Hypothesis (H0):** There is no significant difference in the average sweetness level among the three types of chocolates.

**Alternative Hypothesis (H1):** There is a significant difference in the average sweetness level among the three types of chocolates.

```python
# Sample data (sweetness levels)
chocolate_type_1 = np.array([8, 9, 7, 10, 8, 9, 7, 9, 10, 8])
chocolate_type_2 = np.array([6, 5, 7, 5, 6, 5, 7, 6, 5, 6])
chocolate_type_3 = np.array([10, 11, 9, 12, 11, 12, 10, 11, 12, 10])
```

```python
# Perform the ANOVA test
F_statistic, p_value = stats.f_oneway(chocolate_type_1, chocolate_type_2, chocolate_type_3)

# Define the significance level (alpha) - let's use 0.05
alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is significant difference in average sweetness level among the three types of chocolates.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in the average sweetness level among the three types of chocolates.")
```

# Hypothesis testing using Chi-Square

**Scenario:** A restaurant chain wants to determine if there is a significant association between the type of dish ordered (vegetarian, non-vegetarian, or vegan) and the time of day (breakfast, lunch, or dinner) when customers visit the restaurant.

**Null Hypothesis (H0):** There is no significant association between the type of dish ordered and the time of day.

**Alternative Hypothesis (H1):** There is a significant association between the type of dish ordered and the time of day.

**Data:** The restaurant collected data on a random sample of 200 customers, recording the type of dish ordered and the time of day.

```python
# Contingency table (type of dish vs. time of day)
# Rows: vegetarian, non-vegetarian, vegan
# Columns: breakfast, lunch, dinner
observed_data = np.array([[30, 40, 20],
                          [20, 30, 40],
                          [10, 25, 25]])
```

```python
# Perform the Chi-square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(observed_data)

# Define the significance level (alpha) - let's use 0.05
alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant association between the type of dish ordered and the time of day.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between the type of dish ordered and the time of day.")
```

# Hypothesis testing using correlation coefficient

**Scenario:** A retail company wants to determine if there is a significant correlation between the advertising spending and the sales revenue of their products. They collected data for a sample of 12 months, recording the advertising expenditure (in thousands of dollars) and the corresponding sales revenue (in thousands of dollars) for each month.

**Null Hypothesis (H0):** There is no significant correlation between advertising spending and sales revenue.

**Alternative Hypothesis (H1):** There is a significant correlation between advertising spending and sales revenue.

```python
# Sample data (advertising spending and sales revenue in thousands of dollars)
advertising_spending = np.array([20, 30, 25, 40, 35, 50, 45, 55, 50, 60, 55, 70])
sales_revenue = np.array([40, 60, 50, 70, 65, 80, 75, 85, 80, 90, 85, 100])
```

```python
# Perform the correlation test
correlation_coefficient, p_value = stats.pearsonr(advertising_spending, sales_revenue)

# Define the significance level (alpha) - let's use 0.05
alpha = 0.05

# Check if the p-value is less than the significance level
if p_value < alpha:
    print("Reject the null hypothesis. There is a significant correlation between advertising spending and sales revenue.")
else:
    print("Fail to reject the null hypothesis. There is no significant correlation between advertising spending and sales revenue.")

# Print the correlation coefficient
print("Correlation Coefficient:", correlation_coefficient)
```
