# Limits

```python
from sympy import Symbol, limit, oo, sin, cos, exp, log, sqrt

# Define the variable
x = Symbol('x')

# Basic univariate functions
f1 = x**2 + 3*x - 5
f2 = sin(x) / x
f3 = exp(x) / x
f4 = sqrt(x)
f5 = log(x)

# Compute the limits
limit_f1 = limit(f1, x, 2)
limit_f2 = limit(f2, x, 0)
limit_f3 = limit(f3, x, oo)
limit_f4 = limit(f4, x, 4)
limit_f5 = limit(f5, x, 1)

print("Limit of f1 as x approaches 2:", limit_f1)
print("Limit of f2 as x approaches 0:", limit_f2)
print("Limit of f3 as x approaches infinity:", limit_f3)
print("Limit of f4 as x approaches 4:", limit_f4)
print("Limit of f5 as x approaches 1:", limit_f5)
```

# Continous Function

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x) = x^2
def f(x):
    return x**2

# Define the range of x values for plotting
x_values = np.linspace(-10, 10, 1000)

# Evaluate the function for all x values
y_values = f(x_values)

# Plot the graph
plt.plot(x_values, y_values, label='f(x) = x^2')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of f(x) = x^2')
plt.grid(True)
plt.legend()
plt.show()

# Since f(x) = x^2 is a polynomial function, it is continuous for all real values of x.
print("Continuity Check:")
print("The function f(x) = x^2 is continuous for all real values of x.")

plt.show()
```

# Discontinous Function

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function f(x)
def f(x):
    return np.piecewise(x, [x >= 0, x < 0], [lambda x: 2*x + 1, lambda x: x**2])

# Define the range of x values for plotting
x_values = np.linspace(-5, 5, 1000)

# Evaluate the function for all x values
y_values = f(x_values)

# Plot the graph
plt.plot(x_values, y_values, label='f(x)')
plt.axhline(0, color='gray', lw=0.5)
plt.axvline(0, color='gray', lw=0.5)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Graph of the piecewise function')
plt.grid(True)
plt.legend()
plt.show()

# Check continuity at x = 0
x_0 = 0
limit_right = 2 * x_0 + 1
limit_left = x_0 ** 2

print(f"Continuity/Discontinuity Check:")
print(f"f(0) = {limit_left} (Value at x = 0 from the left side)")
print(f"Limit as x approaches 0 from the right side = {limit_right}")
if limit_right == limit_left:
    print("The function is continuous at x = 0.")
else:
    print("The function is discontinuous at x = 0.")
```

# Applications of Derivatives

## Rate of change of quantity

Find the rate at which a spherical balloon's volume is increasing with respect to its radius, when radius is 7cm.

```python
import math

def volume_of_sphere(radius):
    return (4/3) * math.pi * radius**3

def rate_of_volume_increase(radius):
    return 4 * math.pi * radius**2

def main():
    radius = 7  # Radius in cm
    volume_rate = rate_of_volume_increase(radius)
    print(f"When the radius is {radius} cm, the rate of volume increase is {volume_rate:.2f} cubic centimeters per centimeter.")

if __name__ == "__main__":
    main()
```

## Marginal Cost

Cost function is differentiated to get Marginal Cost.

```python
import sympy as sp

def cost_function(quantity):
    # This function should return the cost for a given quantity.
    return 100 + 5 * quantity + 0.1 * quantity ** 2

def marginal_cost(quantity):
    # Define the symbol for quantity
    q = sp.Symbol('q')
    
    # Define the cost function
    cost_expr = cost_function(q)
    
    # Differentiate the cost function with respect to quantity
    d_cost_expr = sp.diff(cost_expr, q)
    
    # Evaluate the derivative at the given quantity
    mc = d_cost_expr.subs(q, quantity)
    
    return mc

def main():
    quantity_produced = 10
    mc = marginal_cost(quantity_produced)
    print(f"The marginal cost when producing {quantity_produced} units is: ${mc:.2f}")

if __name__ == "__main__":
    main()
```

## Marginal Revenue

Total Revenue is differentiated to get Marginal Revenue.

```python
import sympy as sp

def total_revenue(quantity):
    # This function should return the total revenue for a given quantity.
    return 10 * quantity - 0.1 * quantity ** 2

def marginal_revenue(quantity):
    # Define the symbol for quantity
    q = sp.Symbol('q')

    # Define the total revenue function
    revenue_expr = total_revenue(q)

    # Differentiate the total revenue function with respect to quantity
    d_revenue_expr = sp.diff(revenue_expr, q)

    # Evaluate the derivative at the given quantity
    mr = d_revenue_expr.subs(q, quantity)

    return mr

def main():
    quantity_produced = 5
    mr = marginal_revenue(quantity_produced)
    print(f"The marginal revenue when producing {quantity_produced} units is: ${mr:.2f}")

if __name__ == "__main__":
    main()
```

## Approximation

If radius of a sphere is measured as 9 cm with an error of 0.03cm, then find the approximate error in calculating its volume.

```python
import math

def volume_of_sphere(radius):
    return (4/3) * math.pi * radius**3

def error_in_volume(radius, error_in_radius):
    # Calculate the volume with the measured radius
    V = volume_of_sphere(radius)
    
    # Calculate the derivative of the volume with respect to the radius
    dV_dr = 4 * math.pi * radius**2
    
    # Calculate the approximate error in the volume
    error_in_volume = abs(dV_dr) * error_in_radius
    
    return error_in_volume

def main():
    radius = 9  # Measured radius in cm
    error_in_radius = 0.03  # Error in radius measurement in cm
    
    error_in_volume_approx = error_in_volume(radius, error_in_radius)
    print(f"The approximate error in calculating the volume is: {error_in_volume_approx:.4f} cubic centimeters.")

if __name__ == "__main__":
    main()
```

## Area of Circle

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the variable and the function for the upper half of the circle
x = sp.Symbol('x')
a = 5  # Radius of the circle

# Equation of the circle: x^2 + y^2 = a^2
y_upper = sp.sqrt(a**2 - x**2)

# Calculate the area of one quadrant (upper half) of the circle using integration
area_one_quadrant = sp.integrate(y_upper, (x, 0, a)).evalf()

# Calculate the area of the entire circle by multiplying the area of one quadrant by 4
area_circle = 4 * area_one_quadrant

# Generate x values for the plot (from -a to +a)
x_values = np.linspace(-a, a, 1000)

# Evaluate the upper half of the circle function for the x values
y_values = np.sqrt(a**2 - x_values**2)

# Shade all four quadrants with different colors to complete the circular shape
plt.fill_between(x_values, y_values, color='blue', alpha=0.3)
plt.fill_between(x_values, -y_values, color='green', alpha=0.3)
plt.fill_between(y_values, x_values, color='red', alpha=0.3)
plt.fill_between(-y_values, x_values, color='yellow', alpha=0.3)

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area of Circle with Center (0,0)')
plt.gca().set_aspect('equal', adjustable='box')

# Display the plot
plt.grid()
plt.show()

print(f"The area of the circle with radius {a} and centered at (0,0) is approximately {area_circle:.2f}.")
```

## Area of Parabola

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the variable and the function for the parabola
x = sp.Symbol('x')
a = 2  # Parameter for the parabola

# Equation of the parabola: y^2 = 4ax
y = sp.sqrt(4 * a * x)

# Calculate the area of one quadrant of the parabola using integration
area_one_quadrant = sp.integrate(y, (x, 0, 4/a)).evalf()  # Assuming interval [0, 4/a]

# Calculate the area of the entire parabola by multiplying the area of one quadrant by 2
area_parabola = 2 * area_one_quadrant

# Generate x values for the plot
x_values = np.linspace(0, 4/a, 1000)

# Evaluate the function for the x values
y_values = np.sqrt(4 * a * x_values)

# Shade one quadrant with a color
plt.fill_between(x_values, y_values, alpha=0.3)

# Reflect the parabola about the x-axis to shade the other quadrant with a different color
plt.fill_between(x_values, -y_values, alpha=0.3)

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area Under Parabola')
plt.legend()

# Display the plot
plt.grid()
plt.show()

print(f"The area under the parabola y^2 = 4ax is approximately {area_parabola:.2f}.")
```

## Area between two curves

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the variable and the functions for the curves
x = sp.Symbol('x')
y1 = x**2
y2 = 2*x + 1

# Specify the interval of interest
a = 0
b = 2

# Generate x values for the plot
x_values = np.linspace(a, b, 1000)

# Evaluate the functions for the x values
y1_values = x_values**2
y2_values = 2 * x_values + 1

# Calculate the area between the curves using numerical integration (trapezoidal rule)
area_between_curves = np.trapz(np.abs(y1_values - y2_values), x_values)

# Plot the curves
plt.plot(x_values, y1_values, label='y1 = x^2')
plt.plot(x_values, y2_values, label='y2 = 2x + 1')

# Shade the area between the curves
plt.fill_between(x_values, y1_values, y2_values, where=(y1_values >= y2_values), color='gray', alpha=0.3)
plt.fill_between(x_values, y1_values, y2_values, where=(y1_values < y2_values), color='gray', alpha=0.3)

# Set labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Area Between Curves')
plt.legend()

# Display the plot
plt.grid()
plt.show()

print(f"The area between the curves y1 = x^2 and y2 = 2x + 1 is approximately {area_between_curves:.2f}.")
```

# Graph of Function, Derivative and Integral

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Define the function f(x)
x = sp.Symbol('x')
f = x**3 - 3*x**2 + 2*x + 5

# Calculate the derivative of f(x)
f_prime = sp.diff(f, x)

# Calculate the integral of f(x)
f_integral = sp.integrate(f, x)

# Convert the sympy expressions to lambda functions for numerical evaluation
f_lambda = sp.lambdify(x, f, 'numpy')
f_prime_lambda = sp.lambdify(x, f_prime, 'numpy')
f_integral_lambda = sp.lambdify(x, f_integral, 'numpy')

# Generate x values for plotting
x_vals = np.linspace(-4, 4, 1000)

# Evaluate the functions at the given x values
f_vals = f_lambda(x_vals)
f_prime_vals = f_prime_lambda(x_vals)
f_integral_vals = f_integral_lambda(x_vals)

# Plot the original function, its derivative, and integral
plt.figure(figsize=(10, 6))
plt.plot(x_vals, f_vals, label='f(x) = x^3 - 3x^2 + 2x + 5', color='b')
plt.plot(x_vals, f_prime_vals, label="f'(x) (Derivative)", color='g')
plt.plot(x_vals, f_integral_vals, label="F(x) (Integral)", color='r')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Graph, Derivative, and Integral of f(x)')
plt.grid(True)

plt.show()
```

# Units of Function, Derivative and Integral

```python
import sympy as sp

# Define the function y = 3x^2 + 2x + 5
x = sp.Symbol('x')
y = 3*x**2 + 2*x + 5

# Take the derivative of y with respect to x
dy_dx = sp.diff(y, x)

# Output the derivative
print("Derivative dy/dx =", dy_dx)

# Calculate the units of the derivative
# Note: We assume x is in meters (m) and y is in meters per second (m/s)

# Units of y (output) = m/s
units_y = "m/s"

# Units of x (input) = m
units_x = "m"

# Units of the derivative dy/dx
units_dy_dx = f"{units_y}/{units_x}"  # m/s divided by m

print("Units of the derivative dy/dx:", units_dy_dx)

# Now, let's calculate the integral units of the function y = 3x^2 + 2x + 5
# Assuming x is in meters (m) and y is in meters squared (m^2)

# Take the integral of y with respect to x
integral_y = sp.integrate(y, x)

# Output the integral
print("Integral of y =", integral_y)

# Calculate the units of the integral
# Note: We assume x is in meters (m) and y is in meters squared (m^2)

# Units of y (output) = m^2
units_y_integral = "m^2"

# Units of x (input) = m
units_x_integral = "m"

# Units of the integral ∫y dx
units_integral_y = f"{units_y_integral}*{units_x_integral}"  # m^2 times m

print("Units of the integral ∫y dx:", units_integral_y)
```
